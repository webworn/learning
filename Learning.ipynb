<h3>This code extract float(numeric) value from String Statement</h3>


```python
##Importing re

import re
Statement= "i want 2 pencils 3 pens"
df=re.findall(r"[-+]?(?:\d*\.\d+|\d+)", Statement)
```

```python
print(df)
```

```python
print('Number of Pensils:'+df[0]+" and "+'Number of Pens:'+df[1])
```
